# 딥러닝으로 간단한 수학문제 풀어보기
- 키 = [170, 180, 175, 160]
- 신발 = [260, 270, 265, 255]
- ??? = a*170 + b (그래프에서 선형으로 선을 그릴 수 있다.)

```python
import tensorflow as tf
```

```python
a = tf.Variable(0.1)
b = tf.Variable(0.2)
키 = 170
신발 = 260
```

```python
def 손실함수():
  실제값 = 신발
  예측값 = 키 * a + b
  return tf.square(실제값 - 예측값)
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.1) #learning_rate없어도 됨

for i in range(300):
  opt.minimize(손실함수, var_list=[a, b])# 경사하강법으로 업데이트할 weight Variable 목록)
  print(a.numpy(), b.numpy())
```

```python
pred = 키 * a.numpy() + b.numpy()
print(pred)
```
